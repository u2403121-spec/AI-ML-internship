```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO

# Read the file correctly
with open("test.csv", "r") as f:
    data = f.read()

# Remove extra quotes issue
data = data.replace('"', '')

# Convert to dataframe
df = pd.read_csv(StringIO(data))

# Show first 5 rows
print(df.head())
```

```python
print(df.shape)
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
sns.countplot(x='continent', data=df)

plt.title("Teams by Continent")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(df['fifa_rank_pre_tournament'], bins=10)

plt.title("FIFA Rank Distribution")
plt.xlabel("FIFA Rank")
plt.ylabel("Count")
plt.show()
```

```python
sns.kdeplot(df['goals_scored_last_4y'], fill=True)

plt.title("Goals Scored Distribution")
plt.xlabel("Goals Scored")
plt.show()
```

```python
sns.scatterplot(
    x='wins_last_4y',
    y='losses_last_4y',
    hue='continent',
    data=df
)

plt.title("Wins vs Losses")
plt.show()
```

```python
sns.histplot(df['squad_avg_age'], bins=10)

plt.title("Squad Average Age")
plt.xlabel("Average Age")
plt.show()
```

```python
top_teams = df.sort_values(
    by='squad_total_market_value_eur',
    ascending=False
).head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x='team',
    y='squad_total_market_value_eur',
    data=top_teams
)

plt.xticks(rotation=45)
plt.title("Top 10 Teams by Market Value")
plt.show()
```

```python
continent_counts = df['continent'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    continent_counts,
    labels=continent_counts.index,
    autopct='%1.1f%%'
)

plt.title("Teams by Continent")
plt.show()
```

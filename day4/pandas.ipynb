```python
import pandas as pd
data = {
    "Name": ["John", "Alice", "Bob"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print(df)
```

```python
data = {
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}

df = pd.DataFrame(data)

print(df.groupby("Department").mean())
```

```python
import pandas as pd

data = {
    "Name": ["John", "Alice", "Bob", "Emma", "David"],
    "Math": [90, 85, 78, 92, 88],
    "Science": [95, 80, 85, 89, 91]
}

df = pd.DataFrame(data)

# Add Total Marks
df["Total"] = df["Math"] + df["Science"]

# Add Average
df["Average"] = df["Total"] / 2

# Sort by Total
df = df.sort_values("Total", ascending=False)

print(df)
```
